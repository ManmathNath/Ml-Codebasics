```python
#linear regression-predictef value is continuous value
#logistic regression- predicted value is categorical
#logistic regression is a technique used to solve classification problem
#classification
#when we have only two category as output then it is called binary classification
#when we have more than two category it is called as  multiclass classification
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv("insurance_data.csv")
df
plt.scatter(df["age"],df['bought_insurance'],label="data point")
plt.xlabel("age")
plt.ylabel("bought_insurance")
plt.title("Logistic regression ")
#to draw a linear regression line
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(df[["age"]],df["bought_insurance"])
x= df["age"]
m=reg.coef_
c=reg.intercept_
y_predicted= m* x + c
plt.plot(x,y_predicted,label="LinearRegression",color ="black")
plt.legend()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[["age"]],df['bought_insurance'],random_state=10,train_size=0.9)
#always take the x argument=input data as a 2d array
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)#during fit also take data frames of x argument = input data as 2d array
model.predict(x_test)
model.score(x_test,y_test)
#it predicts the acuracy as 100 percent due to small size of data
model.predict_proba(x_test)
#it produces the the probability for finding whether anyone buy the insurznce or not
#1st term in the class is the buyer don't buy the insurance
```
