```python


#find home price of 3300,5000(in sq.ft)

#predicting homeprice using linear
# regression
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn import linear_model
from matplotlib import style
# to read the csv file
df=pd.read_csv("homeprices.csv")
df
#give names to the axis and plot graph
plt.xlabel("area (sqrft)")
plt.ylabel("price")
plt.title("PRICE PREDICTION MODEL")
plt.scatter(df.area,df.price,marker="*")
plt.grid()
plt.show()
#use y=mx+c,price=slope√óarea+intercept
reg=linear_model.LinearRegression()
reg.fit(df[["area"]],df.price)
reg.predict([[3300]])
reg.predict([[5000]])
#reg.coef_
#reg.intercept_
```

```python
reg.intercept_
Newdf=pd.read_csv("areas.csv")
Newdf
P=reg.predict(Newdf)
Newdf["prices"]= P#i intro a new column inthe  dataframe

Newdf.to_csv("prediction.csv")
Newdf
```
