```python
import pandas as pd
df=pd.read_csv("carprices (1).csv")
df
dummies=pd.get_dummies(df["Car Model"])
dummies
newdf=pd.concat([df,dummies],axis=1)
newdf
final=newdf.drop(["Car Model","Mercedez Benz C class"],axis=1)
final
#before we use train_test_split method we need to define x and y
x=final.drop(["Sell Price($)"],axis=1)
y=df["Sell Price($)"]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.01,random_state=10)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
model.predict(x_test)
model.score(x,y)
#price of a bmw√ó5  7 yr old mileage 86000
model.predict([[86000,7,0,1]])
#price of Mercedez benz 4 yr with 45000 mileage
model.predict([[45000,4,0,0]])
model.score(x,y)
```
