```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn import linear_model
df=pd.read_csv("canada_per_capita_income.csv")
df.head(5)
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.title("per year income  prediction")

plt.scatter(df['year'],df['per capita income (US$)'])
#in plt.scatter sometimes it need to put the colums by the column call
#i.e df['column name '] by this
#bcz when use df.column name it shows errors
#i.e you may forget a comma?
plt.show()
Reg=linear_model.LinearRegression()
Reg.fit(df[["year"]],df['per capita income (US$)'])
Reg.predict([[2020]])
```
